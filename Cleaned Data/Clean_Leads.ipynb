# **II. Into the Features**

```python
id_features = ['Prospect ID', 'Lead Number']
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = df.select_dtypes(include=['object']).columns.tolist()
target = 'Converted'

numerical_features = [col for col in numerical_features if col not in id_features + [target]]
categorical_features = [col for col in categorical_features if col not in id_features + [target]]

print("ID:", id_features)
print("Target:", target)
print("Numerical:", numerical_features)
print("Categorical:", categorical_features)
```

```python
corr_matrix = df[numerical_features].corr()
# Visualize with Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
df = df.drop(['Prospect ID',
              'Lead Number',
              'Asymmetrique Profile Index',
              'Asymmetrique Activity Score',
              'Asymmetrique Activity Index', 
              'Asymmetrique Profile Score', 
              'What matters most to you in choosing a course', 
              'How did you hear about X Education',
              'Lead Quality',
              'Country'], axis=1)
df['What is your current occupation'].fillna('Unknown', inplace=True)
df['Specialization'].fillna('Unknown', inplace=True)
df['City'].fillna('Unknown', inplace=True)
df['Page Views Per Visit'].fillna(df['Page Views Per Visit'].median(), inplace=True)
df['TotalVisits'].fillna(df['TotalVisits'].median(), inplace=True)
df['Last Activity'].fillna('Unknown', inplace=True)
df['Lead Source'].fillna('Unknown', inplace=True)
df['Tags'] = df['Tags'].fillna('Unknown')
df['Lead Profile'] = df['Lead Profile'].fillna('Unknown')
```
